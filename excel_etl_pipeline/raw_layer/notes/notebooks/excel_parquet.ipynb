```python
# Use glob to find all Parquet files in the specified directory
parquet_files = glob.glob(r'../raw_data/output/sales/cash_invoices/summarized/2025/*.parquet')

# Print the list of found Parquet files
print(parquet_files)

# Read all Parquet files into a single DataFrame if any files are found
if parquet_files:
    df = pd.concat([pd.read_parquet(file) for file in parquet_files], ignore_index=True)
else:
    print("No Parquet files found.")
```
