```python
detailed_cash_invoice_path = base_path / "excel_pipeline/data/output/sales/cash_invoices/detailed/2025/01"

print(f"Reading files from: {detailed_cash_invoice_path}")

# Get all parquet files in the directory
detailed_cash_invoice_files = sorted(detailed_cash_invoice_path.glob("*.parquet"))

if detailed_cash_invoice_files:
    # Read and concatenate all parquet files
    detailed_cash_invoice_df = pd.concat(
        [pd.read_parquet(f) for f in detailed_cash_invoice_files], 
        ignore_index=True
    )
    print(f"Found {len(detailed_cash_invoice_files)} parquet files")
    print(f"Total records: {len(detailed_cash_invoice_df)}")
else:
    print(f"No parquet files found in {detailed_cash_invoice_path}")
    detailed_cash_invoice_df = pd.DataFrame()

# Display first few rows
detailed_cash_invoice_df.head()
```

```python

# Read the CSV file
# Using detailed_cash_invoices_jan2025_sample.csv which represents the 'detailed' data type.
file_name = "summarized_cash_invoices_jan2025_sample.csv" 
df = pd.read_csv(file_name)

# Clean up column names by stripping leading/trailing whitespace
df.columns = df.columns.str.strip()

# Display the head and column info
# print("DataFrame Head:")
# print(df.head().to_markdown(index=False, numalign="left", stralign="left"))

print("\nDataFrame Info:")
print(df.info())

# Check for uniqueness of primary key candidate: DOCUMENT_NUMBER
document_number_unique = df['DOCUMENT_NUMBER'].nunique() == len(df)
print(f"\nIs DOCUMENT_NUMBER alone unique? {document_number_unique}")

# Check for uniqueness of primary key candidate: DOCUMENT_NUMBER + ITEM_CODE (Standard line-item PK)
pk_candidates = ['DOCUMENT_NUMBER', 'ITEM_CODE']
pk_check_df = df[pk_candidates].copy()
total_rows = len(df)
unique_combinations = pk_check_df.drop_duplicates()
duplicate_count = total_rows - len(unique_combinations)

print(f"\nPrimary Key Candidate Check ({', '.join(pk_candidates)}):")
print(f"Total Rows: {total_rows}")
print(f"Unique Combinations: {len(unique_combinations)}")
print(f"Duplicate Combinations (Count of rows that are duplicates): {duplicate_count}")

# Show the first few duplicate combinations if they exist
if duplicate_count > 0:
    duplicate_rows = df[pk_check_df.duplicated(keep=False)]
    print("\nSample Duplicate Rows (First 5):")
    print(duplicate_rows.head(5)[pk_candidates].to_markdown(index=False, numalign="left", stralign="left"))
```
